```python
from google.colab import files
uploaded = files.upload()

```

```python
import io
import pandas as pd
import warnings
import numpy as np
warnings.simplefilter("ignore")
df2 = pd.read_excel(io.BytesIO(uploaded['Funil.xlsx']))
df1 = pd.read_excel(io.BytesIO(uploaded['Todas as Usinas.xlsx']))
df2.rename(columns={"Nome": "Usina"}, inplace=True)
df1.index = np.arange(1, len(df1) + 1)
df1['Ranking'] = df1.index
df_merge = pd.merge(df1, df2, on= 'Usina')
df_merge = df_merge.drop(df_merge.columns[[5, 6, 7, 8]], axis=1)
display(df_merge)
```

```python
tam = len(df_merge)


for i in range(tam):
  if df_merge.loc[i, 'Tamanho(ha)'] > 50000:
    df_merge.loc[i, 'Produto'] = 'Produto A'
  elif df_merge.loc[i, 'Tamanho(ha)'] >= 30000 and df_merge.loc[i, 'Tamanho(ha)'] < 50000:
    df_merge.loc[i, 'Produto'] = 'Produto B'
  else:
    df_merge.loc[i, 'Produto'] = 'Produto C'
display(df_merge)
```

```python

df_merge.to_excel('Usinas Atualizado.xlsx', sheet_name= 'atualizado')
```
